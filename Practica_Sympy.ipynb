```c++17
integral_1 = 1/(1+exp(2*x))
integral_1
```

```c++17
integrate(integral_1,x)
```

```c++17
integral_2=exp(x)/sqrt(exp(2*x)+9)
integral_2
```

# Solución de Ecuaciones diferenciales en Sympy #

```c++17
x = sp.Function('x') 
ecu1 = sp.diff(x(t),t)
ecu2 = sp.cos(t**2)

Ecudiff = sp.Eq(ecu1,ecu2)
ci={x(0):0}

Ecudiff

```

```c++17
sol_csi=sp.dsolve(Ecudiff,x(t), ics=ci)
sol_csi
```

### Gráfica ###

```c++17
sp.plot(sol_csi.rhs)
```

```c++17
for i in range(12):
    display(sol_csi.subs(t,i).evalf())
```
